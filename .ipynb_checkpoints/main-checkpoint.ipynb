```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly
```

# Finding null values 

```python
origin = pd.read_csv('data/customer_churn.csv')
df = origin.copy()
df.head()
```

```python
print(f"This DataFrame has: {df.shape[0]} rows and {df.shape[1]} columns")
print(df.info())
print(f"Number of nulls is:\n{df.isnull().sum()}")
```

# No Null Values found, checking for outliers second

```python
df.describe()
```

```python
numeric_cols = ["Call Failure", "Complaints", "Subscription Length", "Charge Amount",
                "Seconds of Use", "Frequency of use", "Frequency of SMS", "Distinct Called Numbers",
                "Age Group", "Tariff Plan", "Status", "Age", "Customer Value", "Churn"]

for col in numeric_cols:
    plt.figure(figsize=(6,3))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(col)
    plt.show()
```

# No outliers detected, starting to ask the questions

## Q1: What percentage of users churn?

```python
churn_percent = df['Churn'].value_counts(normalize=True) * 100
print(f"Percentage of churn users is: {churn_percent[1]:.2f}")
print(f"Percentage of non-churn users is: {churn_percent[0]:.2f}")
# To visualize:
df['Churn'].value_counts().plot(kind="bar", ylabel="percentage", title="Churn distribution")
plt.xticks([0, 1], ['Not Churn', 'Churn'], rotation=0)
```

## Q2: Does usage frequency differ between churned and active users?

```python
usage_churn = df.groupby('Churn')['Frequency of use'].mean()
print("Average Frequency of use:")
print(usage_churn)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Churn', y='Frequency of use', data=df)
```

## Q3: Which age group churns most?

```python
churn_by_age = df.groupby('Age')['Churn'].mean() * 100
print(f"Most churning are {churn_by_age.idxmax()} y/o, with a percentage of {churn_by_age.max():.2f}%")
```

### Verifying which Age belongs to which Age Group

```python
print(df['Age Group'].unique())

df[['Age', 'Age Group']].head(10)
```

### Age Group feature isn't useful because of the Age feature

```python
df = df.drop('Age Group', axis=1)
```

```python
churn_by_age.plot(kind="bar", figsize=(10, 8), color="salmon", title="Churn based on Age", ylabel="Churn Rate (%)")
```

```python
user_count = df['Age'].value_counts().sort_index()
user_count.plot(kind="bar", title='Number of Users per Age', xlabel="Age", ylabel="Number of Users", figsize=(15, 7))
```
